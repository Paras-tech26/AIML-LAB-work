```python
#question 2 (a part)

import numpy as np
import pandas as pd
roll_no=np.arange(1,11)
marks=np.array([78,85,90,67,89,76,98,58,89,63])
attendence=np.array([89,76,78,95,74,59,76,82,51,56])
gender=np.array(['M','F','F',None,'M','M',None,'F','M','F'])
student_df=pd.DataFrame({'roll no':roll_no,
                         'marks':marks,
                         'Attendence':attendence,
                         'Gender':gender
                        })
print(student_df)
```

```python
#question 2 (b part)

null_count=student_df['Gender'].isnull().sum()
print("Null values in gender coloumn is ",null_count)

gender_mode=student_df['Gender'].mode()[0]
student_df['Gender'].fillna(gender_mode,inplace=True)
print(student_df)
```

```python
#question 2 (c part)

avg_marks=student_df['marks'].mean()
print("average marks is ",avg_marks)

gender_count=student_df['Gender'].value_counts()
print(gender_count)
```

```python
#question 3 (a part)

import matplotlib.pyplot as plt
gender_count=student_df['Gender'].value_counts()
plt.bar(gender_count.index,gender_count.values)
plt.xlabel('Gender')
plt.ylabel('number of students')
plt.legend(['students'])
plt.show()
```

```python
#question 3 (b part)

plt.hist(student_df['marks'],bins=5)
plt.xlabel('marks')
plt.ylabel('number of students')
plt.legend(['marks'])
plt.show()
```

```python
#question 3(c part)

plt.hist(student_df['Attendence'],bins=5)
plt.xlabel('Attendence percentage')
plt.ylabel('number of students')
plt.title("Attendence distribution of students")
plt.legend(['Attendence'])
plt.show()
```
