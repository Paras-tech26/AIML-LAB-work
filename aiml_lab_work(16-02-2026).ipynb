```python
#QUESTION given in class


graph = {
    'A': ['B','C'],
    'B': ['D','E','F'],
    'C': ['D','F'],
    'D': ['E','F'],
    'E': ['F'],
    'F': []
}
def dfs(graph, node, visited, goal, path):

    visited.append(node)
    path.append(node)

    if node == goal:
        return True

    for neighbour in graph[node]:
        if neighbour not in visited:
            if dfs(graph, neighbour, visited, goal, path):
                return True

    path.pop()
    return False


start = input("Enter start node: ")
goal = input("Enter goal node: ")

visited = []
path = []

dfs(graph, start, visited, goal, path)

print("DFS Traversal:", visited)
print("Path to goal:", path)
```

```python
#ANOTHER QUESTION GIVEN IN CLASS
#(i) part

from collections import deque
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E', 'F'],
    'C': ['D', 'F'],
    'D': ['E', 'F'],
    'E': ['F'],
    'F': []
}

def bfs(graph, start, goal):
    visited = []
    queue = deque([[start]])

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node not in visited:
            visited.append(node)

            if node == goal:
                return visited, path

            for neighbour in graph[node]:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)

    return visited, []

start = input("Enter start node: ")
goal = input("Enter goal node: ")

traversal, shortest = bfs(graph, start, goal)

print("Traversal Order:", traversal)
print("Shortest Path:", shortest)
```
