```python
#dfs code 1

def dfs(st,goal,graph,vis=None):
    if vis is None:
        vis=set()
    print(st)
    if st==goal:
        return True
    vis.add(st)
    for child in graph[st]:
        if child not in vis:
            if dfs(child,goal,graph,vis):
                return True
D={'A':['B','C'],'B':['A','C'],'C':['A','B','D'],'D':['C','E'],'E':['D']}
dfs('A','E',D,vis=None)
```

```python
#bfs code 1

from collections import deque 
def bfs(st,goal,graph):
    queue=deque([[st]])
    vis=set([st])

    while queue:
        path=queue.popleft()
        node=path[-1]
        if node==goal:
            return path
        for child in graph[node]:
            if child not in vis:
                vis.add(child)
                new_path=path+[child]
                
                queue.append(new_path)
    return none
D={'A':['B','E','D'],'B':['A','E','C'],'C':['B','E','D'],'D':['A','E','C'],'E':['A','B','C','D']}
found=bfs('A','E',D)
print("goal reached",found)
```
